```python
import pandas as pd 
```

```python
df1 = pd.read_json("en_single_choice_test_2K.json", lines=True)
df2 = pd.read_json("en_single_choice_train_3K.json", lines=True)
```

```python
df = pd.concat([df1, df2], ignore_index=True)
print(df.shape)
df.head()
```

```python
df = df[['qid', 'difficulty', 'problem', 'answer_option_list', 'knowledge_point_routes', 'answer_value']]
df.head()
```

```python
def filter_tag(tag):
    # tag is a list, split each element and flatten the result
    if isinstance(tag, list):
        tags = []
        for t in tag:
            tags.extend(t.split("->"))
        # print(tags)
    else:
        tags = tag.split("->")
        # print(tags)

    return tags[-1] if tags else None

data = df.copy()
data['tag'] = data['knowledge_point_routes'].apply(filter_tag)
data.head()

df = data.drop(columns=['knowledge_point_routes'])
df.head()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```
